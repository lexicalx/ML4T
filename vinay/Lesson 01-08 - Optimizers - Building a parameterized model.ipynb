## What is an optimizer?

It is an algorithm that can do the following things:

* Find minimum values of functions Eg., f(x) = x^2+x^3+5 ... Finds the value of x for which f(x) is min
* Build parameterized models based on data Eg., finds a regression function
* Refine allocations to stocks in portfolios

## How to use an optimizer?

* Define a function to minimize
* Provide an initial guess
* Call the optimizer

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as spo

def f(X):
    """Given a scalar X, return some real number value """
    Y = (X - 1.5)**2+0.5
    print "X = {}, Y = {}".format(X,Y)
    return Y

def test_run():
    Xguess = 2.0
    min_result = spo.minimize(f,Xguess,method='SLSQP',options={'disp':True})
    print "Minima found at"
    print "X = {}, Y = {}".format(min_result.x,min_result.fun)
    
    Xplot = np.linspace(0.5,2.5,21)
    Yplot = f(Xplot)
    plt.plot(Xplot,Yplot)
    plt.plot(min_result.x,min_result.fun,'ro')
    plt.title("Minima of an objective function")
    plt.show()
    
if __name__ == "__main__":
    test_run()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as spo

def f(X,M):
    """Given a scalar X, return some real number value """
    Y = ((X - 1.5)**2+0.5)*M
    print "X = {}, Y = {}".format(X,Y)
    return Y

def test_run():
    Xguess = 2.0
    M = 4.0
    min_result = spo.minimize(f,Xguess,args=(M,),method='SLSQP',options={'disp':True})
    print "Minima found at"
    print "X = {}, Y = {}".format(min_result.x,min_result.fun)
    
    Xplot = np.linspace(0.5,2.5,21)
    Yplot = f(Xplot, M)
    plt.plot(Xplot,Yplot)
    plt.plot(min_result.x,min_result.fun,'ro')
    plt.title("Minima of an objective function")
    plt.show()
    
if __name__ == "__main__":
    test_run()
```
